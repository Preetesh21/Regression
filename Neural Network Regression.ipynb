```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
import keras

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, normalize
```

```python
data=pd.read_csv('C:\\Users\\verma\\Desktop\\GPR\\housing.csv')
data.head()
```

```python
#seperating data
datan=np.array(data)
X = datan[:,:3]
y = datan[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
m = 400
f = 3
X_train = X[:m,:f]
X_train = np.c_[np.ones(len(X_train)),X_train]
y_train = y[:m]
X_test = X[m:,:f]
X_test = np.c_[np.ones(len(X_test)),X_test]
y_test = y[m:]
```

```python
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], kernel_initializer='random_normal', activation='relu'))
model.add(Dense(32, kernel_initializer='random_normal', activation='relu'))
model.add(Dense(16, kernel_initializer='random_normal', activation='relu'))
model.add(Dense(8, kernel_initializer='random_normal', activation='relu'))  
model.add(Dense(1, kernel_initializer='random_normal'))
model.compile(loss='mse', optimizer='adam', metrics=['mse'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=170, verbose=1, validation_split=.2)
```

```python
fig = plt.figure(figsize=(13,6))

plt.subplot(2, 2, 1)
plt.plot(history.history['mse'])
plt.title('MSE')
plt.ylabel('Error')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')

plt.subplot(2, 2, 2)
plt.plot(history.history['loss'])
plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')

plt.subplot(2, 2, 3)
plt.plot(history.history['val_loss'])
plt.title('Value Loss')
plt.ylabel('Val Loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper right')

plt.subplot(2, 2, 4)
plt.plot(history.history['val_mse'])
plt.title('Val MSE')
plt.ylabel('Val MSE')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper right')

plt.tight_layout()
plt.show()
plt.tight_layout()
```

```python
y_pred = model.predict(X_test)
mse_pred_score = mean_squared_error(y_pred, y_test)
print('mse_pred_score {}'.format(mse_pred_score))
rmse_pred_score = np.sqrt(mse_pred_score)
print('rmse_pred_score {}'.format(rmse_pred_score))
r2_pred_score = r2_score(y_test, y_pred, multioutput='uniform_average')  
print('r2_pred_score - Coefficient of Determination {}'.format(r2_pred_score))
```

```python
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
fig, ax = plt.subplots(figsize=(12, 10))
plt.style.use('ggplot')
plt.plot(y_pred, y_test, 'ro')
plt.xlabel('Predicted MEDV', fontsize = 15)
plt.ylabel('Actual MEDV', fontsize = 15)
plt.title('Predicted Y (MEDV) to the Actual Y (MEDV)', fontsize = 15)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.show()
```
